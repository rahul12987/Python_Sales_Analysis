# Graphs Creation

**Gender wise count of customers**

```python
gc = sns.countplot(x = 'Gender',data = df, hue='Gender')

for bars in gc.containers:
    gc.bar_label(bars)
```

**Gender wise Total amount of orders**

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

go = sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen, hue='Gender')

for bars in go.containers:
    go.bar_label(bars, fmt=lambda x: f'{x/1000:.0f}K')
```

**Age wise count of customers**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
ax.set_ylabel('Number of customers')
for bars in ax.containers:
    ax.bar_label(bars)
```

**Total amount vs age group**

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ag = sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, hue = 'Age Group')
for bars in ag.containers:
    ag.bar_label(bars, fmt=lambda x: f'{x/1000:.0f}K')
```

**Total number of orders for top 5 States**

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)
sales_state.State

os = sns.barplot(data = sales_state, x = 'State',y= 'Orders', hue = sales_state.State)
os.set_ylabel('Orders Count')

for bars in os.containers:
    os.bar_label(bars)
```

**Total Sales amount for top 5 States**

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)
sales_state

st = sns.barplot(data = sales_state, x = 'State',y= 'Amount', hue = sales_state.State)

for bars in st.containers:
    st.bar_label(bars, fmt=lambda x: f'{x/1000:.0f}K')
```

**Count by Marital Status**

```python
ms_map = {1: 'Married', 0: 'Unmarried'}

# Apply the mapping
df['Ms_Mapped'] = df['Marital_Status'].map(ms_map)

ms = sns.countplot(data = df, x = 'Ms_Mapped', hue = 'Ms_Mapped')
ms.set_xlabel('Marital Status')
for bars in ms.containers:
    ms.bar_label(bars)
```

**Count by Occupation**

```python
co = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')

for bars in co.containers:
    co.bar_label(bars)
```

**Top 5 selling Products**

```python
prod_sales = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(5)

sp = sns.barplot(data = prod_sales, x = 'Product_Category',y= 'Amount', hue = prod_sales.Product_Category)
for bars in sp.containers:
    sp.bar_label(bars, fmt=lambda x: f'{x/1000:.0f}K')
```

**Total Orders from Each Zone**

```python
zone_orders = df.groupby(['Zone'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

zs = sns.barplot(data = zone_orders, x = 'Zone',y= 'Orders', hue = zone_orders.Zone)
for bars in zs.containers:
    zs.bar_label(bars, fmt=lambda x: f'{x/1000:.0f}K')
```

**Percentage of Sales by Zone**

```python
#plt.figure(figsize = (3,4))

zone_sales = df.groupby(['Zone'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

#gb = df.groupby("Zone").agg({'Amount':"Sum"})
plt.pie(zone_sales['Amount'], labels = zone_sales.Zone, autopct = "%1.1f%%")
plt.title("Percentage of Zone vise Sales", fontsize = 10)
plt.show()
```
